```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# ---------------------------
# 1) Create the dataset
# ---------------------------
data = [
    ["Yes","No","No","Compressor"],
    ["Yes","No","No","Compressor"],
    ["Yes","Yes","No","Fan"],
    ["Yes","Yes","Yes","Fan"],
    ["No","Yes","No","Normal"],
    ["No","Yes","Yes","Normal"],
    ["No","No","Yes","Defrost"],
    ["Yes","No","Yes","Defrost"]
]

df = pd.DataFrame(data, columns=["Noise","Cooling","Frost","Issue"])

print("âœ… Dataset:")
print(df)

# ---------------------------
# 2) Convert words to numbers (Encoding)
# ---------------------------
enc_noise = LabelEncoder()
enc_cooling = LabelEncoder()
enc_frost = LabelEncoder()
enc_issue = LabelEncoder()

df["Noise"] = enc_noise.fit_transform(df["Noise"])
df["Cooling"] = enc_cooling.fit_transform(df["Cooling"])
df["Frost"] = enc_frost.fit_transform(df["Frost"])
df["Issue"] = enc_issue.fit_transform(df["Issue"])

# ---------------------------
# 3) Split into X and y
# ---------------------------
X = df[["Noise","Cooling","Frost"]]   # Features (inputs)
y = df["Issue"]                      # Label (output)

# ---------------------------
# 4) Train Decision Tree model
# ---------------------------
model = DecisionTreeClassifier()
model.fit(X, y)

print("\nâœ… Model trained successfully!")

# ---------------------------
# 5) Make 1 prediction (new symptoms)
# Example: Noise=Yes, Cooling=No, Frost=No
# ---------------------------
test_case = pd.DataFrame([[enc_noise.transform(["Yes"])[0],
                           enc_cooling.transform(["No"])[0],
                           enc_frost.transform(["No"])[0]]],
                         columns=["Noise","Cooling","Frost"])

prediction = model.predict(test_case)
predicted_label = enc_issue.inverse_transform(prediction)

print("\nðŸ”® Prediction for Noise=Yes, Cooling=No, Frost=No:")
print("Predicted Issue:", predicted_label[0])
```
