```python
# ----------------------------------------------
# Demo: Python Basics (variables, data structures
# loops, if-else statements)
# ----------------------------------------------

# 1. Variables (different data types)
customer_name = "John Doe"        # string
account_balance = 1200.50         # float
account_active = True             # boolean

print("Customer Name:", customer_name)
print("Account Balance:", account_balance)
print("Account Active:", account_active)
print("-" * 40)

# 2. Data Structures

# List of recent transactions (positive = deposit, negative = withdrawal)
transactions = [200.00, -150.75, -50.00, 500.00, -100.25]

# Dictionary with account details
account_info = {
    "account_number": "ACC12345",
    "account_type": "Savings",
    "interest_rate": 0.03
}

print("Transactions:", transactions)
print("Account Info:", account_info)
print("-" * 40)

# 3. For Loop: Process Transactions
print("Processing Transactions...")

for amount in transactions:
    account_balance += amount
    print("Transaction:", amount, "| Updated Balance:", round(account_balance, 2))

print("-" * 40)

# 4. If-Else: Check Account Status
minimum_balance = 500.00

if account_balance >= minimum_balance:
    print("Account status: Good standing")
else:
    print("Account status: Below minimum balance")

print("-" * 40)

# 5. If-Else Inside a Loop: Categorize Transactions
print("Transaction Summary:")

for amount in transactions:
    if amount > 0:
        print(amount, "is a Deposit")
    else:
        print(amount, "is a Withdrawal")

print("-" * 40)

# 6. Simple Calculation Using Data Structures
total_deposits = 0
total_withdrawals = 0

for amount in transactions:
    if amount > 0:
        total_deposits += amount
    else:
        total_withdrawals += amount

print("Total Deposits:", total_deposits)
print("Total Withdrawals:", abs(total_withdrawals))
print("-" * 40)

# 7. Final Summary
print("Final Account Balance:", round(account_balance, 2))
```
